```python
# import the dependencies
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport as report
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('customer_data.csv').drop('name', axis=1)
X = df.drop('spending', axis=1)
y = df['spending']
df.head()
```

Visualizations, visualizations, and more visualizations!

```python
# Percentage of men vs women in dataset
px.pie(df, names='gender',  title='Ratio of Genders')
```

```python
# Ratio of Degrees
px.pie(df, names='education', title='Ratio of Educations')
```

```python
# Plot Income relative to all other data
for col in ['gender', 'education']:
    px.bar(df, x=col, y='income')
    
```

```python
# Plotting income vs age
fig_age = px.scatter(df, x="income", y="age", color="gender")
fig_age.show()

# Plotting income vs gender
fig_gender = px.box(df, x="gender", y="income")
fig_gender.show()

# Plotting income vs education
fig_education = px.bar(df, x="education", y="income", color="country")
fig_education.show()

# Plotting income vs country
fig_country = px.histogram(df, x="country", y="income", histfunc="sum", marker_color='rgb(102, 205, 170)')
fig_country = fig_country.update_layout(xaxis={'categoryorder':'total descending'})
fig_country.show()

# Plotting income vs purchase_frequency
fig_purchase = px.scatter(df, x="income", y="purchase_frequency", trendline="ols")
fig_purchase.show()

# Plotting income vs spending
fig_spending = px.scatter(df, x="income", y="spending", color="gender", size="age")
fig_spending.show()
```
