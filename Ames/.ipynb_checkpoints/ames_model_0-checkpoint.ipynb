Handle Missing Values

```python
# Sum all of the missing values
null_counts = df.isnull().sum()
null_counts = null_counts[null_counts != 0]
null_counts 
```

For each of the categoric values, we change the null values to a new variable ("U")
For each of the numeric values, we change the null values to the mean ofthe existing data

```python

def transform_values(transformed_df):
    """
    Transform numeric values to the mean of the DataFrame and categorical values to the dummy variable "U".
    """
    # Transform numeric columns to the mean
    numeric_columns = transformed_df.select_dtypes(include=[np.number]).columns
    transformed_df[numeric_columns] = transformed_df[numeric_columns].fillna(transformed_df[numeric_columns].mean())
    
    # Transform categorical columns to "U"
    categorical_columns = transformed_df.select_dtypes(include=['object']).columns
    transformed_df[categorical_columns] = transformed_df[categorical_columns].fillna("U")
    
    return transformed_df

df = transform_values(df)
df
```

```python
# Checking if function worked

null_counts = df.isnull().sum()
null_counts = null_counts[null_counts != 0]

null_counts 
```

Create a function that takes all the categorical variables, maps them to dictionaries, and fills in the dataframe accordingly

```python
def create_categorical_mappings(df):
    categorical_mappings = {}

    for column in df.select_dtypes(include='object'):
        categories = df[column].unique()
        mapping = {category: i+1 for i, category in enumerate(categories)}
        categorical_mappings[column] = mapping
        


    return categorical_mappings

mapping = create_categorical_mappings(df)
mapping
```

```python
# Now apply this to the dataset

def apply_mapping(df, mapping):
    for key, mapping in mapping.items():
        df[key] = df[key].map(mapping)
    
    return df

df = apply_mapping(df, mapping)
df
```

```python
def preprocess(df):
    # Drop ID
    df = df.drop('Id', axis=1)
    return df

df = preprocess(df)
df
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from xgboost import XGBClassifier
from xgboost import XGBRegressor
```

```python
X = df.drop('SalePrice', axis = 1)
y = df["SalePrice"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=69)
```

```python
mlpipe = Pipeline([("initialprocessing", StandardScaler()), ("xgb",XGBClassifier())])
XGBClassifier().get_params()
```

```python
grid = {'xgbclassifier__n_estimators': [1, 10, 100, 500, 1000],
        'booster': [None]}
```

```python
model = GridSearchCV(mlpipe, grid,  cv=5, n_jobs=-1)
```

```python
model.fit(X_train, y_train)
```
